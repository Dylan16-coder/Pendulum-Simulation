# 3D Pendulum motion simulation

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python

def f(a,b):
 return (np.power(b,2)/2-np.cos(a))#Using the equation to determine the motion and returning it 

a=np.linspace(-2*(22/7),2*(22/7),35)#This acts as the theta for the equation
b=np.linspace(-2*(22/7),2*(22/7),35)

A,B=np.meshgrid(a,b)#Creating a meshgrid
C=f(A,B)
 
fig=plt.figure()
ax=plt.axes(projection='3d')
new=ax.plot_surface(B,A,C,rstride=1,cstride=1,cmap='nipy_spectral', edgecolors='black')
fig.colorbar(new)#Creating a colorbar for reference
ax.set_xlabel("theta2")
ax.set_ylabel("theta1")
ax.set_zlabel("motion")
fig.set_size_inches(20,20)#Adjusting the size of 'fig'
```
